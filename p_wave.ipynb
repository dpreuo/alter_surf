```python
# Testing a magnetic field induced p-wave magnet

def H_fct(kx,ky,t=1,m=0, Bx=0, mu=0):
    norbitals = 4 # spin, sublattice
    H = np.zeros((norbitals,norbitals,*kx.shape),dtype=complex)
    H_up = np.zeros((2,2,*kx.shape),dtype=complex)
    H_up[0,0] = - mu
    H_up[1,1] = - mu
    H_up[0,1] = H_up[1,0] = -2*t*cos(kx)

    H[0:2,0:2] = H_up
    H[2:4,2:4] = H_up 
    
    H[0,0] += +m
    H[1,1] += -m
    H[2,2] += -m
    H[3,3] += +m

    H[0,2] += Bx
    H[2,0] += Bx
    H[1,3] += Bx
    H[3,1] += Bx

    return H

H = blochK.Hamiltonian2D(H_fct)
H.add_operator('spin',np.array([1,1,-1,-1]))
H.add_operator('spin_x',np.kron(np.array([[0,1],[1,0]]),np.eye(2)))
H.add_operator('sublattice',np.array([1,-1,1,-1]))

H.set_params(dict(t=1,m=0.5,Bx=0.1,mu=0))

fig,axs = plt.subplots(1,3,figsize=(18,4))
axs[0].set_title('S^z')
plot_bandstruc(axs[0],H, labels_points_path=['-X',r'\Gamma','X'],coloring_operator=H.operator.spin)
axs[1].set_title('S^x')
plot_bandstruc(axs[1],H, labels_points_path=['-X',r'\Gamma','X'],coloring_operator=H.operator.spin_x)
axs[2].set_title('S^z * Sublattice^z')
plot_bandstruc(axs[2],H, labels_points_path=['-X',r'\Gamma','X'],coloring_operator=H.operator.spin*H.operator.sublattice)
```

In the entire script $k_z$ is the out of plane direction and $k_x,k_y$ are in plane. This makes more sense but is rotated with respect to the nature paper. This is implementing the stuff Fernando has on the overleaf

$(k_x,k_y,k_z) \rightarrow (k_x,k_z,-k_y)$

```python
def H_2D_fct(kx,ky,t=1, tz=1, phi=0,m=0):
    norbitals = 4 # spin, sublattice
    H = np.zeros((norbitals,norbitals,*kx.shape),dtype=complex)
    H_up = np.zeros((2,2,*kx.shape),dtype=complex)
    H_up[0,0] = -2*t*(cos(kx)*cos(phi/2)+cos(ky))  - 2*t*sin(kx)*sin(phi/2)
    H_up[1,1] = -2*t*(cos(kx)*cos(phi/2)+cos(ky))  + 2*t*sin(kx)*sin(phi/2)
    H_up[0,1] = H_up[1,0] = -tz

    H[0:2,0:2] = H_up
    H[2:4,2:4] = H_up 
    
    H[0,0] += +m
    H[1,1] += -m
    H[2,2] += -m
    H[3,3] += +m

    return H


def H_slab_fct(kx,ky,len_y=0,t=1,tz=1, phi=0,m=0):
    norbitals = 4 # spin, sublattice
    H = np.zeros((len_y*norbitals,len_y*norbitals,*kx.shape),dtype=complex)

    def H_layer_spin(spin=+1):
        H_up = np.zeros((2,2,*kx.shape),dtype=complex)
        H_up[0,0] = -2*t*(cos(kx)*cos(phi/2))  - 2*t*sin(kx)*sin(phi/2) + spin * m
        H_up[1,1] = -2*t*(cos(kx)*cos(phi/2))  + 2*t*sin(kx)*sin(phi/2) - spin * m
        H_up[0,1] = H_up[1,0] = -tz
        return H_up

    def H_interlayer():
        H_int = np.zeros((4,4,*kx.shape),dtype=complex)
        H_int[0,0] = H_int[1,1] = H_int[2,2] = H_int[3,3] = -t
        return H_int
    

    for j in range(0, len_y):
        H[j*2:(j+1)*2, j*2:(j+1)*2] = H_layer_spin(spin=+1)
        H[(j+1)*2:(j+2)*2, (j+1)*2:(j+2)*2] = H_layer_spin(spin=-1)

        if j>0:
            H[(j-1)*norbitals:j*norbitals, j*norbitals:(j+1)*norbitals] = H_interlayer()
        
    
    H = blochK.utils.hamiltonian_fct.make_hermitian(H)


    return H


H2D = blochK.Hamiltonian2D(H_2D_fct,param=dict(phi=0,m=0),basis=['spin', 'sublattice'], basis_states=['up_A','up_B','down_A','down_B'])
H2D.add_operator('spin', np.array([1,1,-1,-1]))
H2D.add_operator('sublattice', np.array([1,-1,1,-1]))

Hslab = blochK.Hamiltonian2D(H_slab_fct,param=dict(len_y=10,phi=0,m=0),basis=['spin', 'sublattice'], basis_states=['up_A','up_B','down_A','down_B'])
Hslab.add_suboperator('spin', np.array([1,1,-1,-1]))
Hslab.add_suboperator('sublattice', np.array([1,-1,1,-1]))
Hslab.add_operator('spin', np.kron(np.ones(Hslab.param['len_y']),np.array([1,1,-1,-1])))

```

```python
Lk = 51 #number of k-points in each direction

# Define the k-point grid
ks = sample_square(Lk)
#Diagonalize Hamiltonian
H.update_params(dict(PBC=False))
numb_z = H.param['numb_z']
es,psis = H.diagonalize(*ks)


def plot_spectral_density(ax,spectral_density,cmap='bwr',title_plot='',show_xlabel=True,show_ylabel=True,vmax='',vmin='',linthresh=1e-2,alphas=None):
    """Plots colored spin spectral density on ax,

    cmap: a matplotlib colormap
    title_plot: title of the plot if given
    vmax: maximum value for color scaling
    """
    if title_plot!='':
         ax.set_title(title_plot,fontsize=6)
    ax.set_aspect('equal')
    if show_xlabel:
        ax.set_xlabel(r'$k_x$',labelpad=1)
        ax.set_xticks([-pi,0,pi])
        ax.set_xticklabels([r'$-\pi$',0,r'$\pi$'])
    else:
        ax.set_xticklabels([])
    if show_ylabel:
        ax.set_ylabel(r'$k_y$',labelpad=-2.5)
        ax.set_yticks([-pi,0,pi])
        ax.set_yticklabels([r'$-\pi$',0,r'$\pi$'])
    else:
        ax.set_yticklabels([])
    ax.set_xlim(-pi,pi)
    ax.set_ylim(-pi,pi)
    #--------------
    if vmax == '':
        vmax = 1
    if vmin == '':
        vmin = -1
    
    if alphas is None:
        im = ax.imshow(spectral_density.T,extent=(-pi,pi,-pi,pi),origin='lower',cmap=cmap,norm=matplotlib.colors.SymLogNorm(linthresh=linthresh,vmin=vmin,vmax=vmax))
    else:
        im = ax.imshow(spectral_density.T,extent=(-pi,pi,-pi,pi),origin='lower',cmap=cmap,norm=matplotlib.colors.SymLogNorm(linthresh=linthresh,vmin=vmin,vmax=vmax),alpha=alphas.T)
    return im


```

```python
## creating the spectral density data

layers = np.arange(0,H.param['len_z']-1,1,dtype=int)

spectral_densitys = []
spin_spectral_densitys = []
for layer in tqdm(layers):
    proj_spin = H.suboperator.proj.dot(project_doublelayer(layer))*H.operator.spin #projected spin_operator #projected spin_operator

    #compute spin spectral density
    Os = exp_value_O(proj_spin,psis) #.shape=(n,y,x)
    spin_spectral_density = np.sum(Os*Gamma/((es-mu)**2+Gamma**2),axis=0)/H.param['len_z']
    spin_spectral_densitys.append(spin_spectral_density)
    #compute spectral density
    Os = exp_value_O(np.abs(proj_spin),psis) #.shape=(n,y,x)
    spectral_density0 = np.sum(Os*Gamma/((es-mu)**2+Gamma**2),axis=0)/H.param['len_z']
    spectral_densitys.append(spectral_density0)

spectral_densitys = np.array(spectral_densitys)
spin_spectral_densitys = np.array(spin_spectral_densitys)

with open('analysis/spectral_densitys_Weyl.pkl','wb') as f:
    pickle.dump({'spectral_density':spectral_densitys,
                 'spin_spectral_density':spin_spectral_densitys,
                 'layers':layers,
                 'k_mesh':ks}, f)
f.close()

```
